### Enunciado

La primer tarea consistirá en programar una función que genere señales senoidales y que permita parametrizar:

 - la amplitud máxima de la senoidal (volts)
 - su valor medio (volts)
 - la frecuencia (Hz)
 - la fase (radianes)
 - la cantidad de muestras digitalizada por el ADC (# muestras)
 - la frecuencia de muestreo del ADC.

es decir que la función que uds armen debería admitir se llamada de la siguiente manera

tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)

## Función realizada

```python
import numpy as np
import matplotlib.pyplot as plt

#%% Generador de señales senoidales

def     mi_funcion_sen(vmax = 1, dc = 0, ff = 1, ph = 0, nn = 1000, fs = 1000):
    
    w = 2 * np.pi * ff
    tt = np.arange(0, nn/fs, 1/fs)
    xx = dc + vmax * np.sin(tt * w + ph )
    return tt,xx
```

Vamos a probar el funcionamiento de esta funcion:

```python
tt, xx = mi_funcion_sen(1,0,1,0, 1000, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

Ahora vamos a probar modificando sus parametros.

## Modificando la amplitud de la señal 

```python
tt, xx = mi_funcion_sen(5,0,1,0, 1000, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

Podemos observar en el grafico de arriba como se modifico la amplitud de la señal a 5V.

## Modificando el valor medio de la señal

```python
tt, xx = mi_funcion_sen(1,2,1,0, 1000, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

En el grafico podemos observar como la señal fue desplazada con valor medio de 2V

## Modificando frecuencia

```python
tt, xx = mi_funcion_sen(1,0,10,0, 1000, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

En el grafico podemos observar como se modifico la frencuencia pasando de 1Hz a 10Hz.

## Modificando la fase

```python
tt, xx = mi_funcion_sen(1,0,1,np.pi/2, 1000, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

Podemos observar como se modifico la fase 90 grados.

## Modificado la cantidad de muestras

```python
tt, xx = mi_funcion_sen(1,0,1,0, 500, 1000)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

Podemos observar que se pudo graficar la mitad de la señal senoidal porque redujimos la cantidad de muestras a la mitad y afecto al parametro stop = nn/fs 

## Modificando la frecuencia de muestreo

```python
tt, xx = mi_funcion_sen(1,0,1,np.pi/2, 5, 2)

plt.plot(tt, xx)

# Etiquetas y título
plt.xlabel('Tiempo (Seg)')
plt.ylabel('Amplitud (Volt)')
plt.title('Señal generada')

plt.show()
```

Podemos observar en el grafico de arriba como con una fs= 2 Hz el step o paso de muestreo es de 0.5 segundos.
